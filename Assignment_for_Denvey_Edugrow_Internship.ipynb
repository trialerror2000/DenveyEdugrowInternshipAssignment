```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset_3.csv')
df.head()
```

```python
#1. Create a bar chart using matplotlib to compare news sold in each month.
plt.figure(figsize=(10,6))
plt.bar(df['month'], df['news_sold'], color='lightgreen')
plt.title('News Sold Each Month')
plt.xlabel('Month')
plt.ylabel('News Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#2. Add a new column called "percentage_change" that shows the month-to-month percentage change in news sold.
df['percentage_change'] = df['news_sold'].pct_change() * 100
```

```python
#3. Filter and display only those months where news_sold is less than 2000.
filtered_df = df[df['news_sold'] < 2000]
```

```python
#4. 4. Group the months into quarters (Q1, Q2, Q3, Q4) and calculate the total news sold in each quarter.
month_to_quarter = {
    'January': 'Q1', 'February': 'Q1', 'March': 'Q1',
    'April': 'Q2',   'May': 'Q2',      'June': 'Q2',
    'July': 'Q3',    'August': 'Q3',   'September': 'Q3',
    'October': 'Q4', 'November': 'Q4', 'December': 'Q4'
}
df['quarter'] = df['month'].map(month_to_quarter)

quarterly_totals = df.groupby('quarter', sort=False)['news_sold'].sum().reset_index()
quarterly_totals
```

```python
#5. Save only the months with above-average news sold into a new CSV file.
average_news_sold = df['news_sold'].mean()
above_average_df = df[df['news_sold'] > average_news_sold]
above_average_df.to_csv('new.csv', index=False)
```
